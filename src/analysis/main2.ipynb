```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd_problem = pd.read_csv('./data/problem_info.csv')
pd_problem.sort_values(by='contestId', inplace=True)
pd_problem.tail(50)
```

```python
pd_contest = pd.read_csv('./data/cf.csv')
pd_contest['startTimeSeconds'] = pd.to_datetime(
    pd_contest['startTimeSeconds'], unit='s')
pd_contest.rename(columns={'startTimeSeconds': 'startTime'}, inplace=True)
pd_contest
```

```python
pd_contest['type'] = "0"

condition = pd_contest['name'].str.contains(r'\(Div\. 2\)|Rated for Div\. 2', regex=True)
pd_contest.loc[condition, 'type'] = "2"
condition = pd_contest['name'].str.contains('(Div. 1)', regex=False)
pd_contest.loc[condition, 'type'] = "1"
# condition = pd_contest['name'].str.contains('(Div. 3)', regex=False)
# pd_contest.loc[condition, 'type'] = "3"
# condition = pd_contest['name'].str.contains('(Div. 4)', regex=False)
# pd_contest.loc[condition, 'type'] = "4"
condition = pd_contest['name'].str.contains('Div. 1 + Div. 2', regex=False)
pd_contest.loc[condition, 'type'] = "1+2"
pd_contest = pd_contest[pd_contest["type"] != "0"]
merged_df = pd.merge(pd_problem, pd_contest, left_on='contestId', right_on='id')
pd_problem['type'] = merged_df['type_y']
pd_problem['startTime'] = merged_df['startTime']

# 去掉type为Nan的行


pd_problem = pd_problem[pd_problem["index"].notna()]
pd_problem = pd_problem[pd_problem['type'].notna()]
pd_problem = pd_problem[pd_problem['rating'].notna()]

pd_problem['type_index'] = "Div." + pd_problem['type'] + " " + pd_problem['index'].astype(str).str[0]
# print number of row of different type_index
type_index_counts = pd_problem['type_index'].value_counts()
pd_problem = pd_problem.groupby('type_index').filter(lambda x: len(x) > 50)
print(pd_problem['type_index'].value_counts())

# 确保startTime是datetime类型
pd_problem['startTime'] = pd.to_datetime(pd_problem['startTime'])
pd_problem.set_index('startTime', inplace=True)
monthly_ratings = pd_problem.groupby('type_index')['rating'].resample('ME').mean()

# 绘制每个type_index的rating随时间的变化
plt.figure(figsize=(20, 10))
pd_problem = pd_problem.reset_index()
monthly_ratings = monthly_ratings.reset_index()
# 分组绘制
for label, df in monthly_ratings.groupby('type_index'):
    # 确保时间序列的连续性
    full_range = pd.date_range(start=df['startTime'].min(), end=df['startTime'].max(), freq='ME')
    df.set_index('startTime', inplace=True)
    df = df.reindex(full_range, method='ffill')  # 可以选择适当的填充方法，如前向填充

    # 计算累积平均
    df['cumulative_avg'] = df['rating'].expanding().mean()

    # 绘图
    plt.plot(df.index, df['cumulative_avg'], label=label)

plt.title('Cumulative Average Rating Over Time')
plt.xlabel('Time')
plt.ylabel('Cumulative Average Rating')
plt.legend()
plt.show()
print(pd_problem.head())

```

```python

# 假设 pd_problem 已经包含 'type_index' 和 'rating' 列
filtered_data = pd_problem[pd_problem['startTime'] > '2015-01-01']
type_index_avg = filtered_data.groupby('type_index')['rating'].mean().sort_values(ascending=False)

# 打印计算结果
print(type_index_avg)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

pd_problem = pd.read_csv('./data/problem_info.csv')
pd_problem_cleaned = pd_problem.dropna()
merged_df = pd.merge(pd_problem_cleaned, pd_contest, left_on='contestId', right_on='id')
pd_problem_cleaned['startTime'] = merged_df['startTime']

pd_problem_cleaned = pd_problem_cleaned[pd_problem_cleaned['startTime'] > '2018-01-01']

pd_problem_cleaned = pd_problem_cleaned[pd_problem_cleaned['points'] > 100]
pd_problem_cleaned = pd_problem_cleaned[pd_problem_cleaned['rating'] > 100]

print(pd_problem_cleaned.head())

correlation = pd_problem_cleaned['rating'].corr(pd_problem_cleaned['points'])

# 绘制散点图
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rating', y='points', data=pd_problem_cleaned)

# 添加线性回归线
sns.regplot(x='rating', y='points', data=pd_problem_cleaned, scatter=False, color='red')

plt.title(f'Rating vs. Points Correlation (Correlation: {correlation:.2f})')
plt.xlabel('Rating')
plt.ylabel('Points')
plt.grid(True)
plt.show()

# 准备数据
X = pd_problem_cleaned[['points']]  # predictor
y = pd_problem_cleaned['rating']    # response

# 创建线性回归模型并拟合
model = LinearRegression()
model.fit(X, y)

# 预测rating
pd_problem_cleaned['predicted_rating'] = model.predict(X)

# 计算残差
pd_problem_cleaned['residual'] = np.abs(pd_problem_cleaned['rating'] - pd_problem_cleaned['predicted_rating'])

# 找到残差最大的行
worst_correlated_rows = pd_problem_cleaned.nlargest(100, 'residual')  # n 是你想查找的行数

# 显示结果
print(worst_correlated_rows)
```
